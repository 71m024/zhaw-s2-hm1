Teil 1

```python
import numpy as np
from matplotlib import pyplot as plt 
def f(x):
    return x**5 - 5*x**4 - 30*x**3 + 110*x**2 + 29*x -105
a = np.arange(-10, 10, 0.1)
a = f(a)
plt.plot(a)
plt.grid(color='r', linestyle='-', linewidth=1)
plt.xlim(right=180)
plt.xlim(left=40)
plt.ylim(top=120)
plt.ylim(bottom=-120)
plt.show()
```

Die Nullstellen sind bei 50, 90, 110, 130 und 170
