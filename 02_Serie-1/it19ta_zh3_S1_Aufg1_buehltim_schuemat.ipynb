```python
import numpy as np
from matplotlib import pyplot as plt 
def f(x):
    return x**5 - 5*x**4 - 30*x**3 + 110*x**2 + 29*x -105
a = np.arange(-10, 10, 0.1)
a = f(a)
plt.plot(a)
plt.grid(color='r', linestyle='-', linewidth=1)
plt.xlim(right=180)
plt.xlim(left=40)
plt.ylim(top=120)
plt.ylim(bottom=-120)
plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt 

def f(x):
    return x**5 - 5*x**4 - 30*x**3 + 110*x**2 + 29*x -105
def fAbl(x):
    return 5*x**4 - 20*x**3 - 90*x**2 + 220*x + 29
def fAufl(x):
    return 1/6*x**6 - x**5 - 30/4*x**4 + 110/3*x**3 + 29/2*x**2 -105*x

r = np.arange(-10, 10, 0.1)

a = f(r)
b = fAbl(r)
c = fAufl(r)

a_plt, = plt.plot(r, a, label="f(x)")
b_plt, = plt.plot(r, b, label="f'(x)")
c_plt, = plt.plot(r, c, label="F(x)")

plt.grid(color='lightgray', linestyle='-', linewidth=1)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Aufgabe 1")
plt.legend(bbox_to_anchor=(1, 1), handles=[a_plt, b_plt, c_plt])

plt.xlim(right=10)
plt.xlim(left=-8)
plt.ylim(top=1200)
plt.ylim(bottom=-1200)

plt.show()
```
